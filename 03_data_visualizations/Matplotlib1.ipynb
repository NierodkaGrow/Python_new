Automatyzacja powyÅ¼szego kodu

```python
fig = plt.figure(figsize=(12, 10))

for idx, img in enumerate(['office.jpg','business.jpg', 'workplace.jpg', 'children.jpg']):
  plt.subplot(220 + idx + 1)
  plt.imshow(mpimg.imread(img))
```

```python
fig = plt.figure(figsize=(12, 10))

for idx, img in enumerate(['office.jpg','business.jpg', 'workplace.jpg', 'children.jpg']):
  plt.subplot(140 + idx + 1)
  plt.imshow(mpimg.imread(img))
```
